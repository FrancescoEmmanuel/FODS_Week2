```python
import requests
import os
years= list(range(2013,2024))
playerStatsUrl = "https://www.basketball-reference.com/leagues/NBA_{}_per_game.html"

# Create the player directory if it does not exist.
if not os.path.exists("player"):
    os.makedirs("player")

for year in years:
    url = playerStatsUrl.format(year)

    data = requests.get(url)

    with open(f"player/{year}.html", "w+") as f:
        f.write(data.text)
```

```python
from bs4 import BeautifulSoup
import pandas as pd
allYearsDf = []
for year in years:
    with open("player/{}.html".format(year)) as f:
        page = f.read()

    soup = BeautifulSoup(page, 'html.parser')
    soup.find('tr', class_="thead").decompose()
    table = soup.find_all(id="per_game_stats")[0]
    df = pd.read_html(str(table))[0]
    df["Year"] = year
    allYearsDf.append(df)

playerStats = pd.concat(allYearsDf)

```

```python
playerStats
```

```python
playerStats.to_csv("players.csv")
```

```python
teamStatsUrl = "https://www.landofbasketball.com/yearbyyear/2022_2023_standings.htm"
westTeamStats = pd.read_html(teamStatsUrl)[0]
eastTeamStats = pd.read_html(teamStatsUrl)[1]
```

```python
eastTeamStats
```

```python
eastTeamStats = eastTeamStats.dropna(axis = 1)
westTeamStats = westTeamStats.dropna(axis = 1)
```

```python
totalTeamStats = pd.concat([eastTeamStats, westTeamStats])
totalTeamStats
```

```python
totalTeamStats.to_csv("teams.csv")
```
